```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
import numpy
# fix random seed for reproducibility
# load pima indians dataset
df = pd.read_csv("diabetes.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = df.values[:,0:8]
Y = df.values[:,8]


train_x, test_x, train_y, test_y = train_test_split(X,Y,test_size=0.2,random_state=101)
```

```python
# create model
model = Sequential()
model.add(Dense(32, input_dim=train_x.shape[1], activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error',optimizer='adam')
model.fit(train_x,train_y,epochs=150,batch_size=25,verbose=2,validation_data=(test_x,test_y))
```
